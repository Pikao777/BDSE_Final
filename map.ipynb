```python
import folium
import numpy as np

# 建立地圖物件
m = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

# 定義6個有序的座標點
points = [[45.5236, -122.6750], [45.5236, -122.6623], [45.5236, -122.6550],
          [45.5236, -122.6423], [45.5236, -122.6350], [45.5236, -122.6223]]

# 在座標點上添加帶箭頭的自定義標記
for i, point in enumerate(points):
    # 計算箭頭指向的角度
    angle = np.rad2deg(np.arctan2(point[1]-points[i-1][1], point[0]-points[i-1][0]))
    # 使用 folium.RegularPolygonMarker 在座標點上添加帶箭頭的自定義標記
    folium.RegularPolygonMarker(point, popup=str(i+1), fill_color='#132b5e',
                                number_of_sides=3, radius=10, rotation=angle).add_to(m)

# 使用 folium.PolyLine 在標記之間畫線
folium.PolyLine(points, color='red', weight=2.5, opacity=1,
                dash_array='4, 8', popup='Line').add_to(m)

# 顯示地圖
m
```
